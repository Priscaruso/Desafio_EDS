```python
# importa as bibliotecas necess√°rias
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# entrada de dados
# por exemplo: 25/01/2025 25/01/2025 25/02/2025 25/03/2025 25/04/2025

datas = list(input().split())
```

```python
# exibe as datas de entrada
print(datas)
```

```python
# gera o pandas dataframe com a coluna 'data_atendimento'
df = pd.DataFrame({'data_atendimento': pd.to_datetime(datas, dayfirst=True)})
```

```python
# gera dataframe com a quantidade de atendimentos por data de atendimento
df_agrupado = df.groupby('data_atendimento').size().reset_index(name="qtd_atendimentos")
print(df_agrupado)
```

```python
# gr√°fico de linhas de quantidade de atendimentos por dia
plt.plot(df_agrupado['data_atendimento'], df_agrupado['qtd_atendimentos'])
plt.title("Quantidade de atendimentos por dia")
plt.xlabel("Data de atendimento")
plt.ylabel("Quantidade de atendimentos")
plt.xticks(rotation=45)
plt.show()
```

```python
# gr√°fico de barras de quantidade de atendimentos por dia
plt.bar(df_agrupado['data_atendimento'], df_agrupado['qtd_atendimentos'])
plt.title("Quantidade de atendimentos por dia")
plt.xlabel("Data de atendimento")
plt.ylabel("Quantidade de atendimentos")
plt.xticks(rotation=45)
plt.show()
```
